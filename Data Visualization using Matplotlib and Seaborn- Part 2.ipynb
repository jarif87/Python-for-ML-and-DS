```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```
tips=sns.load_dataset("tips")
```

```
tips.head(5)
```

# boxplot

```
plt.boxplot(tips.tip)
plt.text(x=1.1,y=tips["tip"].min(),s="min")
plt.text(x=1.1,y=tips.total_bill.quantile(0.25),s="Q1")
plt.text(x=1.1,y=tips["tip"].median(),s="median (Q2)")
plt.text(x=1.1,y=tips.total_bill.quantile(0.75),s="Q3")
plt.text(x=1.1,y=tips["tip"].max(),s="max")
plt.title("Boxplot")
plt.ylabel("Total Bill")
plt.show()
```

```
plt.boxplot(tips.tip)
plt.text(x=1.1,y=tips["tip"].min(),s="min")
plt.text(x=1.1,y=tips.total_bill.quantile(0.25),s="Q1")
plt.text(x=1.1,y=tips["tip"].median(),s="median (Q2)")
plt.text(x=1.1,y=tips.total_bill.quantile(0.75),s="Q3")
plt.text(x=1.1,y=tips["tip"].max(),s="max")
plt.annotate("Outliers",xy=(0.97,45),xytext=(0.7,44),arrowprops=dict(facecolor="black",arrowstyle="simple"))
plt.title("Boxplot")
plt.ylabel("Total Bill")
plt.show()
```

# seaborn

```
sns.stripplot(y="tip",x="time",data=tips)

plt.show()
```

```
sns.stripplot(y="tip",x="time",data=tips)
sns.stripplot(y=tips["tip"],x=tips["time"])
plt.show()
```

```
sns.swarmplot(y="tip",x="time",data=tips)
plt.show()
```

```
sns.swarmplot(y="tip",x="time",data=tips,hue="day")
plt.show()
```

```
sns.violinplot(y="total_bill",x="day",data=tips)
plt.show()
```

```
sns.violinplot(y="total_bill",x="day",data=tips,hue="sex")
plt.show()
```

```
sns.violinplot(y="total_bill",x="day",data=tips,hue="sex",split=True)
plt.show()
```
