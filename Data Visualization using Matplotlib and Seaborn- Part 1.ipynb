```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
tips=sns.load_dataset("tips")
```

```
tips.head(5)
```

# plot line

```
x=np.linspace(1,20,100)
y=np.exp(x)
plt.plot(x,y)
plt.show()
```

# plot line with color

```
x=np.linspace(1,20,100)
y=np.exp(x)
plt.plot(x,y,"r*")
plt.show()
```

# plot line with different color and parameters

```
x=np.linspace(1,20,100)
y1=x
y2=np.sqrt(x)
y3=np.square(x)

plt.plot(x,y1,"r-",x,y2,"b.",x,y3,"g*")
plt.title("Line Plot\n")
plt.show()
```

# scatter plot

```
tips.head(5)
```

```
x=tips.total_bill
y=tips.tip
```

```
plt.scatter(x,y)
plt.title("Scatter Plot\n")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```
plt.scatter(x,y,s=np.array(y)**2,color="brown",marker="*",alpha=0.5)
plt.title("Scatter Plot\n")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

# bar plot

```
tips.head(5)
```

```
x=tips["smoker"].value_counts()
x.plot(kind="bar",rot=0) #rot=0 returns labels horizontally
plt.show()
```

```
tips["smoker"].value_counts()
```

# bar plot with others parameters

```
x=tips["smoker"].value_counts()
x.plot(kind="bar",rot=0,color="darkred") #rot=0 returns labels horizontally
plt.title("Bar Plot")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()
```

# pie plot

```
tips.head(5)
```

```
data=tips.day.value_counts()
```

```
plt.pie(data,labels=["Sat","Sun","Thur","Fri"],autopct="%.2f%%",radius=1.2)
plt.show()
```

```
plt.pie(data,labels=["Sat","Sun","Thur","Fri"],autopct="%.2f%%",radius=1.2,explode=[0.2,0.3,0.4,0.5])
plt.show()
```

```
explode = (0.05,0.05,0.05,0.05)
plt.pie(data,labels=["Sat","Sun","Thur","Fri"],autopct="%.2f%%",radius=1.2,explode=explode)
#draw circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.tight_layout()
plt.show()
plt.show()
```

```
explode = (0.05,0.05,0.05,0.05)
mycolors = ["gold", "green", "navy", "orangered"]
plt.pie(data,labels=["Sat","Sun","Thur","Fri"],autopct="%.2f%%",radius=1.2,explode=explode,colors=mycolors)
#draw circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.tight_layout()
plt.show()
plt.show()
```
